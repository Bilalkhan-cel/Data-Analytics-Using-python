**What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?**

```python
df['education'].value_counts()
```

```python
total=df.query("education=='Bachelors' or education=='Masters' or education=='Doctorate'")['income'].value_counts().sum()
div=df.query("(education=='Bachelors' or education=='Masters' or education=='Doctorate') and income=='>50K'")['income'].shape[0]
total
print(f"{div/total*100:.2f}%")
```

**What percentage of people with advanced education (Bachelors, Masters, or Doctorate) make more than 50K?**

```python
total=df.query("education!='Bachelors' or education!='Masters' or education!='Doctorate'")['income'].value_counts().sum()
div=df.query("(education!='Bachelors' or education!='Masters' or education!='Doctorate') and income=='>50K'")['income'].shape[0]
total
per=(div/total)*100
print(f"{per:.2f}%")
```

What is the minimum number of hours a person works per week?

```python
df['hours-per-week'].min()
```

What percentage of the people who work the minimum number of hours per week have a salary of more than 50K

```python
min_hours = df['hours-per-week'].min()
total_min_hours_workers = df.query(f"`hours-per-week`=={min_hours}").shape[0]
over_50k_min_hours_workers = df.query(f"`hours-per-week`=={min_hours} and income=='>50K'").shape[0]
percentage = (over_50k_min_hours_workers / total_min_hours_workers) * 100
print(f"{percentage:.2f}%")
```
